```python
df = pd.read_csv("Hotel_Reservations[1].csv")
print(df.shape)
df.head()
```

```python
print(df.info())
df.isna().sum().sort_values(ascending=False).head(10)
```

```python
df = df.drop_duplicates()
```

```python
print(df.isna().sum()) #Se verifican no hay valores nulos que eliminar
```

```python
df['arrival_datetime'] = pd.to_datetime(
    dict(year=df['arrival_year'], month=df['arrival_month'], day=df['arrival_date']),
    errors='coerce'
)  #Se estable una sola columna con la fecha
```

```python
for col in ['type_of_meal_plan','room_type_reserved','market_segment_type','booking_status']:
    df[col] = df[col].str.strip().str.lower()
```

```python
df['total_nights'] = df['no_of_weekend_nights'] + df['no_of_week_nights'] #Creamos nueva varible
```

```python
df['expected_revenue'] = df['avg_price_per_room'] * df['total_nights'] #Ingreso esperado por reserva
```

```python
import sqlite3

# Crear conexin (esto genera un archivo hotel_bookings.db en tu entorno)
conn = sqlite3.connect("hotel_bookings.db")

# Guardar el DataFrame en una tabla llamada 'reservas'
df.to_sql("reservas", conn, if_exists="replace", index=False)

conn.close()
```

```python
conn = sqlite3.connect("hotel_bookings.db")

# Ejemplo: contar reservas por tipo de mercado
query = """
SELECT market_segment_type, COUNT(*) as total_reservas
FROM reservas
GROUP BY market_segment_type
ORDER BY total_reservas DESC;
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT market_segment_type, COUNT(*) AS total_reservas
FROM reservas
GROUP BY market_segment_type
ORDER BY total_reservas DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT room_type_reserved, AVG(avg_price_per_room) AS precio_promedio
FROM reservas
GROUP BY room_type_reserved
ORDER BY precio_promedio DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT arrival_month, SUM(expected_revenue) AS ingresos_totales
FROM reservas
GROUP BY arrival_month
ORDER BY arrival_month;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT market_segment_type, AVG(total_nights) AS estadia_promedio
FROM reservas
GROUP BY market_segment_type
ORDER BY estadia_promedio DESC;
"""
pd.read_sql(query, conn)
```
